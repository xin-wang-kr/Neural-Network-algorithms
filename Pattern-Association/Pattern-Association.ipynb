# Autoassociative Hopfield network

# (a)

```python
import numpy as np

# input training patterns

A_letter = """
.#.
#.#
###
#.#
#.#
"""
A_label = [-1, -1, -1]

D_letter = """
##.
#.#
#.#
#.#
##.
"""
D_label = [-1, 1, 1]

H_letter = """
#.#
#.#
###
#.#
#.#
"""
H_label = [1, 1, 1]

# convert to training matrix and label matrix
def pattern_to_matrix(letter):
    return np.array([+1 if item=='#' else -1 for item in letter.replace('\n','')])

train_set = np.array([pattern_to_matrix(A_letter), pattern_to_matrix(D_letter), pattern_to_matrix(H_letter)])
y_set = np.array([A_label, D_label, H_label])
print(train_set.shape)
print(train_set)
print(y_set.shape)
print(y_set)
```

```python
# Create an autoassociative Hopfield network and train on these three patterns
# Initialize the weight matrix 
W = np.dot(train_set.T, train_set)
print(W.shape)
print(W)
# set W_ij = 0
np.fill_diagonal(W, 0)
W
```

# (b)

```python
# Test the recall on the same three patterns

# do recall
def to_recall(pattern):
    y_i = pattern
    for i in range(0, len(pattern)):
        # Compute net input
        #y_i = pattern
        w_ji = W[:,i]
        y_in_i = pattern[i] + np.dot(y_i, w_ji)
        # Determine activation (threthod = 0, and using bipolar activation function)
        if y_in_i > 0:
            y_out = 1
        if y_in_i < 0:
            y_out = -1
        else:
            y_out = y_i[i]
        y_i[i] = y_out
        print("For n", str(i+1), "Net= ", y_in_i, ", Out= ", y_out, ", upated y_i: ", y_i)
    return y_i

acc_count = 0
for pattern_index in range(0,train_set.shape[0]):
    # Do test for each unit
    print("Test on pattern ", str(pattern_index+1), train_set[pattern_index])
    print("First round")
    y_i_1st = to_recall(train_set[pattern_index])
    print("Second round")
    y_i_2nd = to_recall(y_i_1st)
    # check convergence
    if np.array_equal(y_i_1st, y_i_2nd):
        print(y_i_2nd, " achieves convergence.")
    else:
        print(y_i_2nd, " may need one more iteration.")
    # check if it is successfully recalled
    if np.array_equal(y_i_2nd, train_set[pattern_index]):
        print("Successfully recalled.")
        acc_count += 1
    else:
        print("Not successfully recalled.")
print("Classification accuracy: ", '%.2f' % (acc_count/3))
```

# BAM Network

# (a)

```python
# Create a heteroassociative BAM network and train on these three patterns (15 inputs, and 3 outputs).
# Calculate the weight matrix

A = np.dot(train_set.T, y_set)
print(A.shape)
print(A)
```

# (b)

```python
# Test the recall on the three patterns.

def output(layer, weight_matrix):
    net = np.dot(layer, weight_matrix)
    print("Net: ", net, "Output: ", np.sign(net))
    return np.sign(net)

acc_count = 0
for pattern_index in range(0, train_set.shape[0]):
    print("Test on pattern ", str(pattern_index+1), train_set[pattern_index])
    print("X_layer to Y_layer - 1st round")
    output_y_1st = output(train_set[pattern_index], A)
    print("Y_layer to X_layer - 1st round")
    output_x_1st = output(output_y_1st, A.T)
    print("X_layer to Y_layer - 2nd round")
    output_y_2nd = output(output_x_1st, A)
    print("Y_layer to X_layer - 2nd round")
    output_x_2nd = output(output_y_2nd, A.T)
    # check convergence
    if np.array_equal(output_y_1st, output_y_2nd) and np.array_equal(output_x_1st, output_x_2nd):
        print("X and Y have reached convergence.")
    else:
        print("X and Y may need one more iteration.")
    # check if it is successfully recalled
    #if np.array_equal(output_y_2nd, y_set[pattern_index]):
    if np.array_equal(output_x_2nd, train_set[pattern_index]):
        print("Successfully recalled.")
        acc_count += 1
    else:
        print("Not successfully recalled.")
print("Classification accuracy: ", '%.2f' % (acc_count/3))
```
